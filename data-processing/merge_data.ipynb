```python
import pandas as pd
import glob

csv_files = glob.glob("data/2014/*.csv") 
```

```python
csv_files
```

```python
line_num_sum = 0
for i in range(len(csv_files)):
    with open(csv_files[i]) as f:
        line_num = sum(1 for line in f)
        print(line_num)
        line_num_sum += line_num
print(line_num_sum - len(csv_files) + 1)
```

```python
df = pd.concat((pd.read_csv(f, header = 0) for f in csv_files))
df.to_csv("data/2014.csv")
```
